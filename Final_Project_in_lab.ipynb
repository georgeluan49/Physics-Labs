```
acelerations = np.array([2*coeff_quad1[0], 2*coeff_quad2[0], 2*coeff_quad3[0], 2*coeff_quad4[0], 2*coeff_quad5[0], 2*coeff_quad6[0]])
acelerations
```

```
quad_errs = np.array([np.sqrt(cov_quad1[0,0]), np.sqrt(cov_quad2[0,0]), np.sqrt(cov_quad3[0,0]), np.sqrt(cov_quad4[0,0]), np.sqrt(cov_quad5[0,0]), np.sqrt(cov_quad6[0,0])])
quad_errs
```

```
total_error = np.sqrt(quad_errs[0]**2 + quad_errs[1]**2 + quad_errs[2]**2 + quad_errs[3]**2 + quad_errs[4]**2 + quad_errs[5] **2)/6
total_error
```

```
a = np.mean(acelerations)
a
```

```
I = 0.025 * 0.025 * ((0.250 * 9.81 - 0.250 * a) - (0.215 * 9.81 + 0.215 * a)) / a
I
```

```
v1 = 2 * coeff_quad1[0] * t1 + coeff_quad1[1]
v2 = 2 * coeff_quad2[0] * t2 + coeff_quad2[1]
v3 = 2 * coeff_quad3[0] * t3 + coeff_quad3[1]
v4 = 2 * coeff_quad4[0] * t4 + coeff_quad4[1]
v5 = 2 * coeff_quad5[0] * t5 + coeff_quad5[1]
v6 = 2 * coeff_quad6[0] * t6 + coeff_quad6[1]
```

```
omega1 = (2 * coeff_quad1[0] * t1 + coeff_quad1[1])/0.025
omega2 = (2 * coeff_quad2[0] * t2 + coeff_quad2[1])/0.025
omega3 = (2 * coeff_quad3[0] * t3 + coeff_quad3[1])/0.025
omega4 = (2 * coeff_quad4[0] * t4 + coeff_quad4[1])/0.025
omega5 = (2 * coeff_quad5[0] * t5 + coeff_quad5[1])/0.025
omega6 = (2 * coeff_quad6[0] * t6 + coeff_quad6[1])/0.025
```

```
t = [t1, t2, t3, t4, t5, t6]
p = [p1, p2, p3, p4, p5, p6]
```

```
omega = [omega1, omega2, omega3, omega4, omega5, omega6]
```

```
v = [v1, v2, v3, v4, v5, v6]
```

## Creating the Energy Plots

When I first analyzed the data, I did this process by copying, pasting, and modifying the same code segment. Later, I decided to see if there is a way to automate the process by using a for-loop, and here it is!

```
for i in range(0, 6):
  KE_sensor = 0.5 * 0.215 * (v[i] ** 2)
  KE_weight = 0.5 * 0.250 * (v[i] ** 2)
  PE_sensor = 0.215 * 9.81 * (p[i] - p[i][0])
  PE_weight = 0.250 * 9.81 * (-p[i] + p[i][19])
  KE_pulley = 0.5 * I * (omega[i] ** 2)
  E = KE_sensor + KE_weight + PE_sensor + PE_weight + KE_pulley
  plt.plot(t[i], E, label = 'Total Energy',color = 'black')
  plt.plot(t[i], KE_sensor, label = 'Translational KE$_{sensor}$', color = 'steelblue')
  plt.plot(t[i], KE_weight, label = 'Translational KE$_{weight}$', color = 'red') 
  plt.plot(t[i], PE_sensor, label = 'PE$_{sensor}$', color = 'orange') 
  plt.plot(t[i], PE_weight, label = 'PE$_{weight}$', color = 'purple')
  plt.plot(t[i], KE_pulley, label = 'Rotational KE', color = 'green')
  plt.legend()
  plt.xlabel("Time (s)")
  plt.ylabel("Energy (J)")
  plt.title("Energy Plot Trial "+ str(i+1))
  plt.savefig("Energy_Plot_Trial_"+ str(i+1) +".png",dpi=150)
  plt.show()
```
